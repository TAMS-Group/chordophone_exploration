```python
rospy.init_node('jupyter_lab')
```

```python
import runpy
m= runpy.run_path('../../scripts/run_episode.py')
re=m['RunEpisode']()
```

```python
paths=[]
for x in range(200):
   paths.append(re.get_path_ruckig('a4', direction=-1.0))

sampled_params= np.array([p[1].action_parameters for p in paths])
```
